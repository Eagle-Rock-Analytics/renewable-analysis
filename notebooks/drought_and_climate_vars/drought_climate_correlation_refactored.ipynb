# Renewable Generation and Climate Variable Correlation

This notebook analyzes the relationship between renewable energy generation (via SEI) and climate variables at specific geographic locations.

**Key Features:**
- Point-based analysis at user-specified lat/lon location
- Flexible SEI source: compute from renewable data or load pre-computed file
- Flexible climate metrics: anomalies or raw values
- Multiple climate variables: temperature, wind, precipitation, humidity
- Analysis across models and global warming levels
- Scatterplots overlaying historical and future GWL periods

**Data Sources:**
- Renewable generation data: S3 bucket zarr files
- Climate data: WRF downscaled via climakitae
- Pre-computed SEI: Optional NetCDF files from previous analyses

```python
import numpy as np
import pandas as pd
import xarray as xr
import matplotlib.pyplot as plt
import os
import sys
from pathlib import Path

# Import holoviews for interactive plotting
import holoviews as hv
from holoviews import opts
hv.extension('bokeh')

# Import custom modules
sys.path.append('../../src')
from renewable_data_load import sim_name_dict, get_gwl_crossing_period, get_ren_data_concat
from plotting_config import model_colors, gwl_colors, model_markers
from climate_correlation import (
    load_wrf_climate_data_all_simulations,
    extract_simulation_gwl_period,
    extract_point_data,
    find_nearest_valid_point,
    compute_anomalies,
    compute_generation_sei_at_location,
    load_generation_sei_from_file,
)

# Load matplotlib style sheet
plt.style.use('../../renewable_analysis.mplstyle')


print("Imports successful!")
```

## Configuration

Set analysis parameters including:
- Location (lat/lon or grid indices)
- SEI data source (compute or load from file)
- Climate variable to analyze
- Whether to use anomalies or raw values
- Models and GWL periods

```python
# =============================================================================
# LOCATION SPECIFICATION
# =============================================================================
# Option 1: Provide lat/lon (will find nearest valid grid cell)
USE_LATLON = True
target_lat = 33.9  # San Bernardino area
target_lon = -116.6

# Option 2: Provide grid indices directly (set USE_LATLON = False to use this)
x_idx = 66
y_idx = 116

# =============================================================================
# SEI DATA SOURCE
# =============================================================================
# Option 1: Compute SEI from renewable generation data
COMPUTE_SEI = True  # Set to True to compute SEI on-the-fly

# Parameters for SEI computation (only used if COMPUTE_SEI = True)
resource = "pv"  # Options: "pv", "windpower"
module = "utility"  # Options: "utility" (for pv), "onshore" or "offshore" (for windpower)
domain = "d02"  # Options: "d02" (9km), "d03" (3km)
variable = "cf"  # capacity factor
frequency = "day"
sei_window_size = 30  # Rolling window size for SEI calculation (days)

# Option 2: Load pre-computed SEI from file (only used if COMPUTE_SEI = False)
# sei_filepath = "../../data/location_analysis/sei_ec-earth3_pv_utility_d02.nc"

# =============================================================================
# CLIMATE VARIABLE
# =============================================================================
climate_variable = "Air Temperature at 2m"  # Options:
# "Air Temperature at 2m"
# "Mean wind speed at 10m"
# "Precipitation (total)"
# "Relative humidity"

# =============================================================================
# ANALYSIS CONFIGURATION
# =============================================================================
# Climate data configuration
climate_use_anomaly = True  # Calculate anomalies for climate variable

# Model and GWL selection
simulation = 'ec-earth3'  # Options: "ec-earth3", "miroc6", "mpi-esm1-2-hr", "taiesm1"
reference_gwl = 0.8  # Historical reference warming level
future_gwl = 2.0  # Future warming level to compare

# Output directory
output_dir = Path("./figures/climate_correlation")
output_dir.mkdir(parents=True, exist_ok=True)

print("Configuration complete!")
```

## Determine Grid Location

If using lat/lon, find the nearest valid grid cell. This shows which grid cell will be used for the analysis.

```python
if COMPUTE_SEI and USE_LATLON:
    print("Finding nearest valid grid cell for target location...")
    print(f"  Target: ({target_lat:.2f}°N, {target_lon:.2f}°E)")
    print(f"  Loading sample renewable data to find valid grid cells...")
    
    # Load a sample of renewable data to find valid grid cells
    sample_data = get_ren_data_concat(
        resource=resource,
        module=module,
        domain=domain,
        variable=variable,
        frequency=frequency,
        simulation=simulation,
    ).isel(time=0)  # Just need one time slice to check valid locations
    
    # Find nearest valid point
    x_idx, y_idx = find_nearest_valid_point(sample_data, target_lat, target_lon)
    
    print(f"\n✓ Will use grid cell at x={x_idx}, y={y_idx}")
    
else:
    if not COMPUTE_SEI:
        print("Skipping grid location determination (using pre-computed SEI)")
    else:
        print(f"Using provided grid indices: x={x_idx}, y={y_idx}")
```

## Load or Compute SEI Data

Either compute SEI from renewable generation data or load from a pre-computed file.

```python
if COMPUTE_SEI:
    print(f"\nComputing SEI from renewable generation data...")
    print(f"  Resource: {resource} - {module}")
    print(f"  Domain: {domain}")
    print(f"  Simulation: {simulation}")
    print(f"  Grid location: x={x_idx}, y={y_idx}")
    
    # Always use grid indices (already determined from lat/lon if needed)
    sei_data = compute_generation_sei_at_location(
        resource=resource,
        module=module,
        domain=domain,
        simulation=simulation,
        x_idx=x_idx,
        y_idx=y_idx,
        reference_gwl=reference_gwl,
        window_size=sei_window_size,
        variable=variable,
        frequency=frequency,
    )
    
    print(f"\n✓ SEI computed successfully!")
    print(f"  Shape: {sei_data.shape}")
    print(f"  Dimensions: {sei_data.dims}")
    
else:
    print(f"Loading pre-computed SEI from file...")
    print(f"  File: {sei_filepath}")
    
    sei_data = load_generation_sei_from_file(sei_filepath, convert_to_timeseries=True)
    
    print(f"\nSEI loaded successfully!")
    print(f"  Shape: {sei_data.shape}")
    print(f"  Dimensions: {sei_data.dims}")
```

## Load Climate Data

Load WRF climate variable and extract for the same location.

```python
print(f"Loading climate data: {climate_variable}")
print("  This loads all simulations at once (efficient single call)...")

# Load all simulations with one call
climate_data_all_sims = load_wrf_climate_data_all_simulations(
    variable=climate_variable,
    domain=domain,
    trim_edges=True
)

print(f"  ✓ Loaded: {climate_data_all_sims.dims}")
print(f"  Available simulations: {list(climate_data_all_sims.simulation.values)[:4]}...")
```

## Extract Simulation and GWL Periods

Extract the specific simulation and crop to GWL periods, then extract point location.

```python
print(f"\nExtracting {simulation} for GWL periods...")

# Dictionary to store climate data by GWL
climate_by_gwl = {}

for gwl in [reference_gwl, future_gwl]:
    print(f"\n  GWL {gwl}°C:")
    
    # Extract simulation and GWL period
    climate_gwl = extract_simulation_gwl_period(
        climate_data=climate_data_all_sims,
        simulation=simulation,
        gwl=gwl
    )
    
    print(f"    Time period: {climate_gwl.time.values[0]} to {climate_gwl.time.values[-1]}")
    
    # Extract point location
    if USE_LATLON:
        climate_point = extract_point_data(
            climate_gwl,
            lat=target_lat,
            lon=target_lon
        )
    else:
        climate_point = extract_point_data(
            climate_gwl,
            x_idx=x_idx,
            y_idx=y_idx
        )
    
    climate_by_gwl[gwl] = climate_point
    print(f"    Point extracted: {climate_point.shape} points")

print("\n✓ Climate data extraction complete!")
```

## Compute Anomalies (if requested)

Calculate day-of-year anomalies for climate variable if configured.

```python
if climate_use_anomaly:
    print("Computing climate anomalies...")
    print(f"  Using GWL {reference_gwl}°C as reference period")
    
    # Save reference climatology
    reference_climate = climate_by_gwl[reference_gwl].copy()
    
    # Compute anomalies for both GWLs using the reference period
    for gwl in [reference_gwl, future_gwl]:
        climate_by_gwl[gwl] = compute_anomalies(
            climate_by_gwl[gwl],
            reference_data=reference_climate
        )
    
    print("  ✓ Anomalies computed relative to reference period")
    climate_label = f"{climate_variable} Anomaly"
else:
    print("Using raw climate values (no anomalies)")
    climate_label = climate_variable
```

## Extract SEI for GWL Periods

Crop SEI data to match the GWL periods from climate data.

```python
print("Extracting SEI for GWL periods...")

# Dictionary to store SEI data by GWL
sei_by_gwl = {}

for gwl in [reference_gwl, future_gwl]:
    # Get time range from climate data
    start_time = climate_by_gwl[gwl].time.values[0]
    end_time = climate_by_gwl[gwl].time.values[-1]
    
    # Extract SEI for this period
    sei_gwl = sei_data.sel(time=slice(start_time, end_time))
    sei_by_gwl[gwl] = sei_gwl
    
    print(f"  GWL {gwl}°C: {len(sei_gwl.time)} days")

print("✓ SEI extraction complete!")
```

## Align Data for Correlation Analysis

**Critical Step:** Use `xr.align()` to ensure SEI and climate data are matched on exact dates.
This prevents silent misalignment that could corrupt correlation analysis.

```python
print("Aligning SEI and climate data on time coordinates...")

# Aligned data dictionaries
sei_aligned = {}
climate_aligned = {}

for gwl in [reference_gwl, future_gwl]:
    # Explicitly align on time dimension with inner join
    sei_gwl_aligned, climate_gwl_aligned = xr.align(
        sei_by_gwl[gwl],
        climate_by_gwl[gwl],
        join='inner'  # Only keep dates present in both datasets
    )
    
    sei_aligned[gwl] = sei_gwl_aligned
    climate_aligned[gwl] = climate_gwl_aligned
    
    n_points = len(sei_gwl_aligned.time)
    print(f"  GWL {gwl}°C: {n_points} aligned data points")
    
    # Check if alignment removed any data
    if n_points != len(sei_by_gwl[gwl].time):
        n_removed = len(sei_by_gwl[gwl].time) - n_points
        print(f"    Warning: {n_removed} dates removed due to mismatch")

print("\n✓ Data alignment complete!")
```

## Create Correlation Scatterplot

Generate interactive scatterplot comparing SEI vs climate variable, overlaying historical and future GWL periods.

```python
def create_sei_climate_scatterplot(
    sei_gwl1,
    climate_gwl1,
    sei_gwl2,
    climate_gwl2,
    simulation,
    climate_label,
    gwl1=0.8,
    gwl2=2.0,
    overlay=True
):
    """
    Create scatterplot of SEI vs climate variable for two GWLs.
    
    Parameters
    ----------
    sei_gwl1, sei_gwl2 : xr.DataArray
        SEI data for each GWL (already aligned)
    climate_gwl1, climate_gwl2 : xr.DataArray
        Climate data for each GWL (already aligned)
    simulation : str
        Model name
    climate_label : str
        Label for climate axis
    gwl1, gwl2 : float
        Warming levels
    overlay : bool
        If True, overlay both GWLs on single plot
        
    Returns
    -------
    holoviews plot
    """
    # Prepare data for GWL1
    df_gwl1 = pd.DataFrame({
        'sei': sei_gwl1.values.flatten(),
        'climate': climate_gwl1.values.flatten(),
        'time': sei_gwl1.time.values.flatten()
    })
    df_gwl1['GWL'] = f'{gwl1}°C'
    
    # Prepare data for GWL2
    df_gwl2 = pd.DataFrame({
        'sei': sei_gwl2.values.flatten(),
        'climate': climate_gwl2.values.flatten(),
        'time': sei_gwl2.time.values.flatten()
    })
    df_gwl2['GWL'] = f'{gwl2}°C'
    
    if overlay:
        # Overlaid plot - combine data and color by GWL
        df_combined = pd.concat([df_gwl1, df_gwl2], ignore_index=True)
        
        scatter = hv.Scatter(df_combined, 'climate', ['sei', 'GWL']).opts(
            width=700, height=600, size=5, alpha=0.1,
            color='GWL',
            cmap={f'{gwl1}°C': gwl_colors.get(gwl1, 'blue'), 
                  f'{gwl2}°C': gwl_colors.get(gwl2, 'red')},
            tools=['hover', 'box_select'],
            xlabel=climate_label,
            ylabel='Generation SEI',
            title=f'{simulation.upper()} | GWL {gwl1}°C vs {gwl2}°C',
            show_legend=False,
        )
        
        # Add invisible points for legend
        points_legend = hv.Scatter(
            pd.concat([df_combined.head(5), df_combined.tail(5)]),
            'climate', ['sei', 'GWL']
        ).opts(
            size=0, alpha=1.0,
            color='GWL',
            cmap={f'{gwl1}°C': gwl_colors.get(gwl1, 'blue'), 
                  f'{gwl2}°C': gwl_colors.get(gwl2, 'red')},
            show_legend=True,
            legend_position='top_right'
        )
        
        # Add reference lines
        hline = hv.HLine(0).opts(color='black', line_width=1, line_dash='dashed', alpha=0.5)
        vline = hv.VLine(0).opts(color='black', line_width=1, line_dash='dashed', alpha=0.5)
        
        return scatter * points_legend * hline * vline
    
    else:
        # Side-by-side plots
        scatter_gwl1 = hv.Scatter(df_gwl1, 'climate', 'sei').opts(
            width=450, height=400, size=5, alpha=0.1,
            color=gwl_colors.get(gwl1, 'blue'),
            tools=['hover', 'box_select'],
            xlabel=climate_label,
            ylabel='Generation SEI',
            title=f'GWL {gwl1}°C'
        )
        
        scatter_gwl2 = hv.Scatter(df_gwl2, 'climate', 'sei').opts(
            width=450, height=400, size=5, alpha=0.1,
            color=gwl_colors.get(gwl2, 'red'),
            tools=['hover', 'box_select'],
            xlabel=climate_label,
            ylabel='Generation SEI',
            title=f'GWL {gwl2}°C'
        )
        
        # Add reference lines
        hline = hv.HLine(0).opts(color='black', line_width=1, line_dash='dashed', alpha=0.5)
        vline = hv.VLine(0).opts(color='black', line_width=1, line_dash='dashed', alpha=0.5)
        
        return (scatter_gwl1 * hline * vline + scatter_gwl2 * hline * vline).opts(
            shared_axes=True
        )
```

### Generate Final Scatterplot

```python
# Create overlaid scatterplot
plot = create_sei_climate_scatterplot(
    sei_aligned[reference_gwl],
    climate_aligned[reference_gwl],
    sei_aligned[future_gwl],
    climate_aligned[future_gwl],
    simulation=simulation,
    climate_label=climate_label,
    gwl1=reference_gwl,
    gwl2=future_gwl,
    overlay=True
)

plot
```

### Side-by-Side Comparison

Alternative visualization showing GWLs separately.

```python
# Create side-by-side scatterplot
plot_sidebyside = create_sei_climate_scatterplot(
    sei_aligned[reference_gwl],
    climate_aligned[reference_gwl],
    sei_aligned[future_gwl],
    climate_aligned[future_gwl],
    simulation=simulation,
    climate_label=climate_label,
    gwl1=reference_gwl,
    gwl2=future_gwl,
    overlay=False
)

plot_sidebyside
```

## Summary Statistics

Calculate correlation coefficients and basic statistics.

```python
print("Correlation Analysis Summary")
print("="*60)
print(f"Simulation: {simulation.upper()}")
print(f"Climate Variable: {climate_variable}")
if COMPUTE_SEI:
    print(f"SEI Source: {resource} - {module} (computed)")
else:
    print(f"SEI Source: Pre-computed file")
print(f"Location: {sei_aligned[reference_gwl].lat.values:.2f}°N, {sei_aligned[reference_gwl].lon.values:.2f}°E")
print()

for gwl in [reference_gwl, future_gwl]:
    # Calculate Pearson correlation
    corr = np.corrcoef(
        sei_aligned[gwl].values.flatten(),
        climate_aligned[gwl].values.flatten()
    )[0, 1]
    
    print(f"GWL {gwl}°C:")
    print(f"  Correlation coefficient: {corr:.3f}")
    print(f"  Number of data points: {len(sei_aligned[gwl].time)}")
    print(f"  SEI mean: {sei_aligned[gwl].mean().values:.3f}")
    print(f"  SEI std: {sei_aligned[gwl].std().values:.3f}")
    print(f"  Climate mean: {climate_aligned[gwl].mean().values:.3f}")
    print(f"  Climate std: {climate_aligned[gwl].std().values:.3f}")
    print()

print("="*60)
```
