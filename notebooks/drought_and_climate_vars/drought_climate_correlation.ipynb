```python
resource = "pv"
module = "utility"
domain = "d02"
variable = "cf"
frequency = "day"
reference_gwl = 0.8
future_gwl = 2.0
simulations = ["mpi-esm1-2-hr", "miroc6", "taiesm1", 'ec-earth3',]
simulation = "taiesm1"

target_gwl = future_gwl

wrf_var = "Precipitation (total)"
#wrf_var = "Air Temperature at 2m"
short_name = 'prec'
```

```python
# Select a single location
x_idx, y_idx = 66, 116

#210, 231
# san bernadino 66, 116
```

```python
ren_ds = get_ren_data_gwl(resource, module, domain, variable, frequency, simulation, [reference_gwl, future_gwl])
ren_ds_ref = ren_ds[0].isel(x=x_idx, y=y_idx)
ren_ds_fut = ren_ds[1].isel(x=x_idx, y=y_idx)

ren_ds_ref_copy = ren_ds_ref[0].copy()
ren_ds_ref_copy['time'] = ren_ds_fut[0].time

ren_anom_fut = ren_ds_fut[0].groupby('time.dayofyear') - ren_ds_ref_copy.groupby('time.dayofyear').mean('time', skipna=True)
ren_anom_ref = ren_ds_ref[0].groupby('time.dayofyear') - ren_ds_ref[0].groupby('time.dayofyear').mean('time', skipna=True)


# get WRF data
wrf_simulation = sim_name_dict[simulation]
wrf_scenario = ['Historical Climate', 'SSP 3-7.0']
wrf_resolution = resolution_dict[domain]
wrfdat = get_data(
    variable=wrf_var,
    downscaling_method="Dynamical",
    resolution=wrf_resolution,
    timescale="daily",
    scenario=wrf_scenario,
)
wrfdat = wrfdat.sel(simulation=wrf_simulation).isel(x=x_idx, y=y_idx)
wrfdat = wrfdat.convert_calendar("noleap")

# crop the time axis of the WRF data to match the ren data
start_time_ref = ren_anom_fut.time.values[0]
end_time_ref = ren_anom_fut.time.values[-1]
wrfdat_ref = wrfdat.sel(time=slice(start_time_ref, end_time_ref))

start_time_fut = ren_anom_fut.time.values[0]
end_time_fut = ren_anom_fut.time.values[-1]
wrfdat_fut = wrfdat.sel(time=slice(start_time_fut, end_time_fut))

wrf_anom_ref = wrfdat_ref.groupby('time.dayofyear') - wrfdat_ref.groupby('time.dayofyear').mean('time', skipna=True)
wrf_anom_fut = wrfdat_fut.groupby('time.dayofyear') - wrfdat_ref.groupby('time.dayofyear').mean('time', skipna=True)


# merge the ren and wrf data on time, x, y
merged_ds_ref = xr.merge([ren_anom_ref, wrf_anom_ref], join='inner')
merged_ds_fut = xr.merge([ren_anom_fut, wrf_anom_fut], join='inner')
#merged_ds = merged_ds.rename_vars({"Air Temperature at 2m": var})

var = wrf_var
```

```python


# Create a tap stream with default location
tap = Tap(source=img, x=210, y=231)

# Define a dynamic callback that updates the scatter plot based on tap location
def create_tap_plot(x, y):
    if x is None or y is None:
        x, y = 210, 231
    
    # Round to nearest integer
    x_idx = int(round(x))
    y_idx = int(round(y))
    
    # Bounds check
    x_idx = max(0, min(x_idx, merged_ds.dims['x']-1))
    y_idx = max(0, min(y_idx, merged_ds.dims['y']-1))
    
    # Extract data
    location_data = merged_ds.isel(x=x_idx, y=y_idx).squeeze()
    df = location_data.to_dataframe().reset_index()
    
    # Create scatter plot
    scatter = hv.Scatter(df, var, 'cf').opts(
        width=450, height=400, size=5, alpha=0.3, color='blue',
        tools=['hover', 'box_select'],
        xlabel='Temperature Anomaly (Â°C)', 
        ylabel='CF Anomaly',
        title=f'Location: x={x_idx}, y={y_idx}'
    )
    
    # Add regression line
    slope, intercept = np.polyfit(df[var], df['cf'], 1)
    x_line = np.array([df[var].min(), df[var].max()])
    y_line = slope * x_line + intercept
    line = hv.Curve((x_line, y_line)).opts(color='red', line_width=2, line_dash='dashed')
    
    return scatter * line

# Create dynamic map
dmap = hv.DynamicMap(create_tap_plot, streams=[tap])

# Display side by side
(img + dmap).opts(shared_axes=False)
```

```python
def latlon_to_xy(lat_target, lon_target, data=base_map_data):
    """
    Find the nearest non-NaN grid cell to the target lat/lon coordinates.
    
    Parameters
    ----------
    lat_target : float
        Target latitude in degrees
    lon_target : float
        Target longitude in degrees
    data : xr.DataArray
        Spatial data with lat/lon coordinates (default: base_map_data)
    
    Returns
    -------
    x_idx : int
        X index of nearest valid grid cell
    y_idx : int
        Y index of nearest valid grid cell
    """
    # Get lat/lon coordinates from the data
    lats = data.lat.values
    lons = data.lon.values
    
    # Calculate distance to all grid cells
    distances = np.sqrt((lats - lat_target)**2 + (lons - lon_target)**2)
    
    # Mask out NaN values in the data
    valid_mask = ~np.isnan(data.isel(warming_level=0).values)
    distances_masked = np.where(valid_mask, distances, np.inf)
    
    # Find the indices of the minimum distance
    min_idx = np.unravel_index(np.argmin(distances_masked), distances_masked.shape)
    y_idx, x_idx = min_idx
    
    return int(x_idx), int(y_idx)
```
