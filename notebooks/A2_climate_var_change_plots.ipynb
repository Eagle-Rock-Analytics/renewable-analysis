```python
import numpy as np
import pandas as pd

import cartopy.crs as ccrs
import cartopy.feature as cfeature
import matplotlib.pyplot as plt

import xarray as xr
import matplotlib.pyplot as plt

import os 
import time


# Import holoviews and set up for interactive plotting
import holoviews as hv
from holoviews import opts
hv.extension('bokeh')

from renewable_data_load import *

from climakitae.core.data_interface import (
    get_data)
```

```python
simulations = ["mpi-esm1-2-hr", "miroc6", "taiesm1", 'ec-earth3']
simulation = "taiesm1"
domain = "d02"
wrf_var = "Precipitation (total)"
```

```python
# get WRF data
wrf_simulation = sim_name_dict[simulation]
wrf_scenario = ['Historical Climate', 'SSP 3-7.0']
wrf_resolution = resolution_dict[domain]
wrfdat = get_data(
    variable=wrf_var,
    downscaling_method="Dynamical",
    resolution=wrf_resolution,
    timescale="monthly",
    approach='warming level',
    warming_level=[0.8,2.0]
)
#wrfdat = wrfdat.sel(simulation=wrf_simulation).isel(x=x_idx, y=y_idx)
#wrfdat = wrfdat.convert_calendar("noleap")
```

```python
wrf_sims = [
       'WRF_EC-Earth3_r1i1p1f1_historical+ssp370',
       'WRF_MIROC6_r1i1p1f1_historical+ssp370',
       'WRF_MPI-ESM1-2-HR_r3i1p1f1_historical+ssp370',
       'WRF_TaiESM1_r1i1p1f1_historical+ssp370']

wrfdat = wrfdat.sel(simulation=wrf_sims)
```

```python
delta = wrfdat.mean('time_delta')
delta = delta.sel(warming_level=2.0) - delta.sel(warming_level=0.8)
```

```python
delta.nbytes / 1e9
```

```python
delta = delta.compute()
```

```python
fig, axes = plt.subplots(
    1, 4, figsize=(13, 3), subplot_kw={"projection": ccrs.PlateCarree()}
)
for ax, wrf_sim in zip(axes.flatten(), wrf_sims, strict=False):
    delta.sel(simulation=wrf_sim).plot(
        ax=ax,
        x="lon",
        y="lat",
        transform=ccrs.PlateCarree(),
        cbar_kwargs={"label": None},

        cmap="RdBu",
        vmin=-30,
        vmax=30

    )
    ax.set_title(wrf_sim)
    ax.set_extent([-130, -95, 25, 57])
    ax.add_feature(cfeature.STATES.with_scale("10m"), edgecolor="black", linewidth=0.1)
    ax.add_feature(cfeature.COASTLINE.with_scale("10m"), edgecolor="black")
# set title
fig.suptitle(f"Change in {wrf_var} (2.0°C - 0.8°C)", fontsize=14)
#plt.savefig(f"figures/large_scale_cf_change_{resource}_{module}_{gwl_fut}C_minus_{gwl_ref}C.png", dpi=300, bbox_inches='tight')
```
