# Climate Drivers of Electricity Demand

This notebook analyzes the relationship between electricity demand and climate variables (temperature, wind, precipitation) across California load zones.

**Key Features:**
- Regional analysis using load zone shapefiles
- Flexible demand metrics: SEI (standardized anomaly) or raw peak load
- Flexible climate metrics: anomalies or raw values
- Multiple climate variables: temperature, wind speed, precipitation
- Analysis across models and global warming levels

**Data Sources:**
- Demand SEI: `compute_demand_sei.ipynb` → `data/SEI/`
- Demand peak load: `data/demand/DailyPeaks/`
- Climate data: WRF downscaled via climakitae
- Regional boundaries: `data/load_zone_shapefiles/allLoadZones.shp`

```python
import numpy as np
import pandas as pd
import xarray as xr
import matplotlib.pyplot as plt
import os
import sys
from pathlib import Path

# Import holoviews for interactive plotting
import holoviews as hv
from holoviews import opts
hv.extension('bokeh')

# Import custom modules
sys.path.append('../../src')
from renewable_data_load import sim_name_dict, get_gwl_crossing_period
from plotting_config import model_colors, gwl_colors, model_markers
from climate_correlation import (
    load_wrf_climate_data_all_simulations,
    extract_simulation_gwl_period,
    compute_regional_climate_average,
    compute_anomalies,
    prepare_correlation_data,
    load_demand_sei,
    load_demand_peak_load
)

# Load matplotlib style sheet
plt.style.use('../../renewable_analysis.mplstyle')

print("Imports successful!")
```

## Configuration

Set analysis parameters including:
- Climate variables to analyze
- Demand data type (SEI or raw peak load)
- Whether to use anomalies or raw values
- Models and GWL periods
- Output directory

```python
# Analysis parameters
simulations = ["ec-earth3", "miroc6", "mpi-esm1-2-hr", "taiesm1"]
gwls = [0.8, 2.0]  # Reference and future warming levels
domain = "d02"

# Climate variables to analyze
climate_variables = {
    "Air Temperature at 2m": "T2m",
    "Relative humidity":"rh"
    # "Mean wind speed at 10m": "Wind10m",
   #"Precipitation (total)": "Precip"
}

# Demand data configuration
demand_data_type = "peak_load"  # Options: "sei" or "peak_load"
demand_use_anomaly = False  # Only applies if demand_data_type="peak_load"

# Climate data configuration
climate_use_anomaly = False  # Calculate anomalies for climate variables

# Shapefile path
shapefile_path = "../../data/load_zone_shapefiles/allLoadZones.shp"

# Output directory
output_dir = Path("./figures/climate_drivers_of_demand")
output_dir.mkdir(parents=True, exist_ok=True)
```

## Load Demand Data

Load demand data (SEI or raw peak load) for all simulations and regions.

```python
# Load demand data for all simulations
demand_data = {}

if demand_data_type == "sei":
    print("Loading demand SEI data...")
    for simulation in simulations:
        try:
            demand_data[simulation] = load_demand_sei(
                simulation=simulation,
                data_dir="../../data/SEI",
                domain=domain,
                reference_gwl=0.8
            )
            print(f"  {simulation}: {demand_data[simulation].dims}")
        except FileNotFoundError as e:
            print(f"  Warning: {e}")
else:  # peak_load
    print("Loading raw peak load data...")
    for simulation in simulations:
        try:
            # Load full timeseries (will crop by GWL later)
            demand_data[simulation] = load_demand_peak_load(
                simulation=simulation,
                data_dir="../../data/demand/DailyPeaks"
            )
            print(f"  {simulation}: {demand_data[simulation].dims}")
        except Exception as e:
            print(f"  Warning: Could not load {simulation}: {e}")

print(f"\nLoaded demand data for {len(demand_data)} simulations")

# Show available regions
if demand_data:
    first_sim = list(demand_data.keys())[0]
    print(f"Available regions: {list(demand_data[first_sim].region.values)}")
```

## Load and Process Climate Data

Load WRF climate variables and aggregate to regional averages using load zone shapefiles.


```python
# Dictionary to store regional climate data
# Structure: regional_climate[simulation][gwl][climate_var_short_name] = xr.DataArray
regional_climate = {}

# Initialize nested structure
for simulation in simulations:
    regional_climate[simulation] = {}
    for gwl in gwls:
        regional_climate[simulation][gwl] = {}

# Load each climate variable ONCE (efficient!)
print("Loading climate data (one get_data call per variable)...")
print("="*60)

climate_data_cache = {}

for climate_var, var_short in climate_variables.items():
    print(f"\n{climate_var}:")
    print(f"  Loading all simulations... (this may take a minute)")
    
    try:
        # SINGLE call to get_data - returns all 8 simulations
        climate_data_all_sims = load_wrf_climate_data_all_simulations(
            variable=climate_var,
            domain=domain,
            trim_edges=True
        )
        
        climate_data_cache[var_short] = climate_data_all_sims
        
        print(f"  ✓ Loaded: {climate_data_all_sims.dims}")
        print(f"  Available simulations: {list(climate_data_all_sims.simulation.values)}")
        
    except Exception as e:
        print(f"  ✗ Error loading {climate_var}: {e}")
        continue

print("\n" + "="*60)
print(f"Climate data loaded for {len(climate_data_cache)} variables")
print("="*60)
```

## Extract Simulations and Aggregate to Regions

Now extract each simulation/GWL combination and aggregate to load zones.

```python
# Process each simulation/GWL combination
print("\nExtracting simulations and aggregating to regions...")
print("="*60)

for simulation in simulations:
    print(f"\n{simulation}:")
    
    for gwl in gwls:
        print(f"  GWL {gwl}°C:")
        
        for var_short, climate_data_all in climate_data_cache.items():
            try:
                # Extract this simulation and GWL period
                climate_data = extract_simulation_gwl_period(
                    climate_data=climate_data_all,
                    simulation=simulation,
                    gwl=gwl
                )
                
                # Aggregate to regions
                regional_data = compute_regional_climate_average(
                    climate_data=climate_data,
                    shapefile_path=shapefile_path
                )
                
                # Store
                regional_climate[simulation][gwl][var_short] = regional_data
                
                print(f"    {var_short}: ✓")
                
            except Exception as e:
                print(f"    {var_short}: ✗ Error: {e}")
                continue

print("\n" + "="*60)
print("Regional climate data processing complete!")
print("="*60)
```

## Compute Anomalies (if requested)

Calculate day-of-year anomalies for climate variables if configured.

```python
# Process climate anomalies if requested
if climate_use_anomaly:
    print("Computing climate anomalies...")
    print("  Using GWL 0.8 as reference period for all warming levels")
    
    reference_gwl = 0.8
    
    for simulation in regional_climate:
        # First, save the reference period data (GWL 0.8) BEFORE computing anomalies
        reference_data_saved = {}
        for var_short in regional_climate[simulation][reference_gwl]:
            # Make a copy to preserve original data
            reference_data_saved[var_short] = regional_climate[simulation][reference_gwl][var_short].copy()
        
        # Now compute anomalies for all GWLs using the saved reference
        for gwl in regional_climate[simulation]:
            for var_short in regional_climate[simulation][gwl]:
                # Compute anomalies relative to GWL 0.8 climatology
                regional_climate[simulation][gwl][var_short] = compute_anomalies(
                    regional_climate[simulation][gwl][var_short],
                    reference_data=reference_data_saved[var_short]
                )
    
    print("Anomalies computed relative to GWL 0.8 reference period")
else:
    print("Using raw climate values (no anomalies)")
```

## Create Correlation Scatterplots

Generate interactive scatterplots comparing demand vs climate variables.

Each plot shows:
- Demand (y-axis): SEI or raw peak load
- Climate variable (x-axis): anomaly or raw value
- Points colored by GWL
- Separate plots for each model, region, and climate variable

**Data Alignment (Critical):**
- Uses `xr.align()` with `join='inner'` to ensure demand and climate data are matched on exact dates
- Only includes dates present in both datasets
- Prevents silent misalignment that could corrupt correlation analysis

```python
def create_demand_climate_scatterplot(
    demand_data_gwl1, 
    climate_data_gwl1,
    demand_data_gwl2,
    climate_data_gwl2,
    simulation,
    region,
    climate_var_name,
    gwl1=0.8,
    gwl2=2.0,
    demand_label="Demand SEI",
    climate_label="Climate Variable",
    overlay=False
):
    """
    Create scatterplots for two GWLs (side-by-side or overlaid).
    
    Parameters
    ----------
    demand_data_gwl1 : xr.DataArray
        Demand data for first GWL
    climate_data_gwl1 : xr.DataArray
        Climate data for first GWL
    demand_data_gwl2 : xr.DataArray
        Demand data for second GWL
    climate_data_gwl2 : xr.DataArray
        Climate data for second GWL
    simulation : str
        Model name
    region : str
        Region name
    climate_var_name : str
        Climate variable name for labels
    gwl1, gwl2 : float
        Warming levels
    demand_label : str
        Label for demand axis
    climate_label : str
        Label for climate axis
    overlay : bool
        If True, overlay both GWLs on single plot. If False, side-by-side plots.
        
    Returns
    -------
    holoviews.Layout or holoviews.Overlay
        Side-by-side scatter plots or overlaid scatter plot
    """
    # Explicitly align data on time coordinates (CRITICAL for accurate correlation)
    # This ensures demand and climate are matched exactly on date
    aligned_demand_gwl1, aligned_climate_gwl1 = xr.align(
        demand_data_gwl1, 
        climate_data_gwl1, 
        join='inner'  # Only keep dates present in both datasets
    )
    
    aligned_demand_gwl2, aligned_climate_gwl2 = xr.align(
        demand_data_gwl2, 
        climate_data_gwl2, 
        join='inner'
    )
    
    # Prepare data for GWL1 - now guaranteed to be aligned
    df_gwl1 = pd.DataFrame({
        'demand': aligned_demand_gwl1.values.flatten(),
        'climate': aligned_climate_gwl1.values.flatten(),
        'time': aligned_demand_gwl1.time.values.flatten()
    })
    
    # Prepare data for GWL2
    df_gwl2 = pd.DataFrame({
        'demand': aligned_demand_gwl2.values.flatten(),
        'climate': aligned_climate_gwl2.values.flatten(),
        'time': aligned_demand_gwl2.time.values.flatten()
    })
    
    # Add GWL label to both dataframes for overlaid legend
    df_gwl1['GWL'] = f'{gwl1}°C'
    df_gwl2['GWL'] = f'{gwl2}°C'
    
    # Log alignment info for validation
    n_points_gwl1 = len(df_gwl1)
    n_points_gwl2 = len(df_gwl2)
    if n_points_gwl1 != len(demand_data_gwl1.time) or n_points_gwl2 != len(demand_data_gwl2.time):
        print(f"  Note: Alignment removed mismatched dates. GWL1: {n_points_gwl1} points, GWL2: {n_points_gwl2} points")
    
    if overlay:
        # Overlaid plot - combine data and color by GWL
        df_combined = pd.concat([df_gwl1, df_gwl2], ignore_index=True)
        
        scatter = hv.Scatter(df_combined, 'climate', ['demand', 'GWL']).opts(
            width=600, height=500, size=5, alpha=0.08, muted_alpha=1,
            color='GWL',
            cmap={f'{gwl1}°C': gwl_colors.get(gwl1, 'blue'), 
                  f'{gwl2}°C': gwl_colors.get(gwl2, 'red')},
            tools=['hover', 'box_select'],
            xlabel=climate_label,
            ylabel=demand_label,
            title=f'{region} | {simulation}\nGWL {gwl1}°C vs {gwl2}°C',
            show_legend=False,
        )


        points_legend = hv.Scatter(pd.concat([df_combined.head(10),df_combined.tail(10)]), 'climate', ['demand', 'GWL']).opts(
            size=0,                   # Invisible on plot
            alpha=1.0,                # Fully opaque for legend
            color='GWL',
            cmap={f'{gwl1}°C': gwl_colors.get(gwl1, 'blue'), 
                f'{gwl2}°C': gwl_colors.get(gwl2, 'red')},
            show_legend=True,         # This drives the legend
            legend_position='top_right'
)
        scatter = scatter * points_legend



        return scatter
    else:
        # Side-by-side plots
        scatter_gwl1 = hv.Scatter(df_gwl1, 'climate', 'demand').opts(
            width=450, height=400, size=5, alpha=0.1,
            color=gwl_colors.get(gwl1, 'blue'),
            tools=['hover', 'box_select'],
            xlabel=climate_label,
            ylabel=demand_label,
            title=f'{region} | {simulation}\nGWL {gwl1}°C'
        )
        
        scatter_gwl2 = hv.Scatter(df_gwl2, 'climate', 'demand').opts(
            width=450, height=400, size=5, alpha=0.1,
            color=gwl_colors.get(gwl2, 'red'),
            tools=['hover', 'box_select'],
            xlabel=climate_label,
            ylabel=demand_label,
            title=f'{region} | {simulation}\nGWL {gwl2}°C'
        )
        
        # Return side-by-side layout
        return (scatter_gwl1 + scatter_gwl2).opts(shared_axes=True)
```

## Generate Plots for All Combinations

Loop through all simulations, regions, and climate variables to create scatterplots.

**Plot Layout Options:**
- **Side-by-side** (`overlay_plots=False`): Shows two separate plots for easy direct comparison
- **Overlaid** (`overlay_plots=True`): Shows both GWLs on one plot with color-coded legend for pattern comparison

```python
# Select a subset for initial exploration
test_simulation = "taiesm1"  # Change to explore different models

# Plot configuration
overlay_plots = True  # True = overlay both GWLs on one plot, False = side-by-side

# Set labels based on configuration
if demand_data_type == "sei":
    demand_label = "Demand SEI"
elif demand_use_anomaly:
    demand_label = "Demand Anomaly (MW)"
else:
    demand_label = "Peak Demand (MW)"


climate_label_suffix = " Anomaly" if climate_use_anomaly else ""
```

```python
# Example: Temperature vs Demand for EC-Earth3, PGE region
simulation = test_simulation
region = "SCE"
var_short = "T2m"
climate_var_full = "Air Temperature at 2m"

gwl1, gwl2 = gwls[0], gwls[1]

print(f"\nPlotting {var_short} vs Demand")
print(f"  Simulation: {simulation}")
print(f"  Region: {region}")
print(f"  GWLs: {gwl1}°C, {gwl2}°C")

# Get demand data for this region and GWLs
if demand_data_type == "sei":
    # Get the time periods for GWL filtering
    wrf_sim_name = sim_name_dict[simulation]
    model = wrf_sim_name.split("_")[1]
    ensemble_member = wrf_sim_name.split("_")[2]
    
    start_year_1, end_year_1 = get_gwl_crossing_period(model, ensemble_member, gwl1)
    start_year_2, end_year_2 = get_gwl_crossing_period(model, ensemble_member, gwl2)
    
    # Select region
    demand_gwl1 = demand_data[simulation].sel(region=region)
    demand_gwl2 = demand_data[simulation].sel(region=region)
    
    # Check if data has 'year' and 'dayofyear' dimensions or just 'time'
    if 'year' in demand_gwl1.dims and 'dayofyear' in demand_gwl1.dims:
        # SEI data with year/dayofyear structure
        demand_gwl1 = demand_gwl1.sel(year=slice(start_year_1, end_year_1))
        demand_gwl2 = demand_gwl2.sel(year=slice(start_year_2, end_year_2))
        
        # Stack back to time dimension for plotting
        demand_gwl1_stacked = demand_gwl1.stack(time=['year', 'dayofyear']).reset_index('time')
        demand_gwl2_stacked = demand_gwl2.stack(time=['year', 'dayofyear']).reset_index('time')
    else:
        # SEI data with regular time dimension
        demand_gwl1_stacked = demand_gwl1.sel(
            time=slice(f"{start_year_1}-01-01", f"{end_year_1}-12-31")
        )
        demand_gwl2_stacked = demand_gwl2.sel(
            time=slice(f"{start_year_2}-01-01", f"{end_year_2}-12-31")
        )
    
else:
    # Raw peak load
    demand_gwl1 = load_demand_peak_load(simulation, gwl=gwl1).sel(region=region)
    demand_gwl2 = load_demand_peak_load(simulation, gwl=gwl2).sel(region=region)
    
    if demand_use_anomaly:
        demand_gwl1 = compute_anomalies(demand_gwl1)
        demand_gwl2 = compute_anomalies(demand_gwl2)
    
    demand_gwl1_stacked = demand_gwl1
    demand_gwl2_stacked = demand_gwl2

# Get climate data
climate_gwl1 = regional_climate[simulation][gwl1][var_short].sel(region=region)
climate_gwl2 = regional_climate[simulation][gwl2][var_short].sel(region=region)

# Create plot
plot = create_demand_climate_scatterplot(
    demand_gwl1_stacked,
    climate_gwl1,
    demand_gwl2_stacked,
    climate_gwl2,
    simulation=simulation,
    region=region,
    climate_var_name=climate_var_full,
    gwl1=gwl1,
    gwl2=gwl2,
    demand_label=demand_label,
    climate_label=climate_var_full + climate_label_suffix,
    overlay=overlay_plots
)
filename = f"{simulation}_{region}_{var_short}_vs_demand_{demand_data_type}.html"
filepath = output_dir / filename
hv.save(plot, filepath)
plot
```

## Correlation Statistics

Calculate correlation coefficients between demand and climate variables.

```python
# Calculate correlations for all combinations
correlation_results = []

for simulation in simulations:
    if simulation not in demand_data or simulation not in regional_climate:
        continue
    
    regions = demand_data[simulation].region.values
    
    for region in regions:
        for gwl in gwls:
            for climate_var_full, var_short in climate_variables.items():
                try:
                    # Prepare demand data
                    if demand_data_type == "sei":
                        demand = demand_data[simulation].sel(region=region)
                        
                        # Filter by GWL period
                        wrf_sim_name = sim_name_dict[simulation]
                        model = wrf_sim_name.split("_")[1]
                        ensemble_member = wrf_sim_name.split("_")[2]
                        start_year, end_year = get_gwl_crossing_period(model, ensemble_member, gwl)
                        
                        # Check dimensions and filter appropriately
                        if 'year' in demand.dims and 'dayofyear' in demand.dims:
                            # SEI with year/dayofyear structure
                            demand = demand.sel(year=slice(start_year, end_year))
                            demand_stacked = demand.stack(time=['year', 'dayofyear']).reset_index('time')
                        else:
                            # SEI with regular time dimension
                            demand_stacked = demand.sel(
                                time=slice(f"{start_year}-01-01", f"{end_year}-12-31")
                            )
                    else:
                        demand = load_demand_peak_load(simulation, gwl=gwl).sel(region=region)
                        if demand_use_anomaly:
                            demand = compute_anomalies(demand)
                        demand_stacked = demand
                    
                    # Get climate data
                    climate = regional_climate[simulation][gwl][var_short].sel(region=region)
                    
                    # CRITICAL: Explicitly align on time coordinate before correlation
                    aligned_demand, aligned_climate = xr.align(
                        demand_stacked,
                        climate,
                        join='inner'  # Only dates present in both
                    )
                    
                    # Extract values from aligned data
                    demand_vals = aligned_demand.values.flatten()
                    climate_vals = aligned_climate.values.flatten()
                    
                    # Remove NaNs
                    mask = ~(np.isnan(demand_vals) | np.isnan(climate_vals))
                    demand_vals = demand_vals[mask]
                    climate_vals = climate_vals[mask]
                    
                    # Calculate correlation
                    if len(demand_vals) > 0:
                        corr = np.corrcoef(demand_vals, climate_vals)[0, 1]
                        
                        correlation_results.append({
                            'simulation': simulation,
                            'region': region,
                            'gwl': gwl,
                            'climate_var': climate_var_full,
                            'correlation': corr,
                            'n_points': len(demand_vals)
                        })
                
                except Exception as e:
                    continue

# Convert to DataFrame
corr_df = pd.DataFrame(correlation_results)
print(f"Calculated {len(corr_df)} correlations")
corr_df.head(20)
```

```python
# Summary statistics by climate variable and GWL
if len(corr_df) > 0:
    summary = corr_df.groupby(['climate_var', 'gwl'])['correlation'].agg(['mean', 'std', 'min', 'max'])
    print("\nCorrelation Summary:")
    print(summary)
```
